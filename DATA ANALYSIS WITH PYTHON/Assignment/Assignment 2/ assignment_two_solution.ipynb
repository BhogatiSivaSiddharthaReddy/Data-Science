```python
import openpyxl
```

```python
path = "Retail_Sample.xlsx"
```

```python
wb_obj = openpyxl.load_workbook(path)
sheet_obj = wb_obj.active
```

```python
row = sheet_obj.max_row
column = sheet_obj.max_column
```

```python
print(row)
print(column)
```

```python

dates_list = list()
for i in range(2, row + 1):
    cell_obj = sheet_obj.cell(row = i, column = 1)
    dates_list.append(cell_obj.value)
unique_dates = set(dates_list)
print("No.of unique dates in the list is: ",len(unique_dates)) #answer to "no of unique dates"
unique_months = list()
for i in dates_list:
    m = i.month
    if m in unique_months:
        pass
    else:
        unique_months.append(m)
print("unique months in the data set are :",sorted(unique_months))
```

```python
revenue_dict = dict()
for i in range(2, row + 1):
    dates = str(sheet_obj.cell(row = i, column = 1).value)
    quantity = sheet_obj.cell(row = i, column = 3).value
    unit_price = sheet_obj.cell(row = i, column = 5).value
    total_price = quantity * unit_price
    if dates in revenue_dict:
        revenue_dict[dates] = float(revenue_dict.get(dates)) + float(total_price)
    else:
        revenue_dict[dates] = float(total_price)
print(revenue_dict)

```

```python
month_revenue_dict = dict()
for i in range(2, row + 1):
    month = sheet_obj.cell(row = i, column = 1).value.month
    quantity = sheet_obj.cell(row = i, column = 3).value
    unit_price = sheet_obj.cell(row = i, column = 5).value
    total_price = quantity * unit_price
    if month in month_revenue_dict:
        month_revenue_dict[month] = float(month_revenue_dict.get(month)) + float(total_price)
    else:
        month_revenue_dict[month] = float(total_price)
print("Highest and least quantity for each month is:  ",month_revenue_dict)
```
