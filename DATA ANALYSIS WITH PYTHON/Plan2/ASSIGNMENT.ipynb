```python
"""
    problem: wikipedia page of their favorite celebrity 
             collect first 3 paragraphs in text
             
    output: - create a dictionary with words as keys and their counts as values
            1. sorted by keys  2. sorted by values
            
            - percentage of articles for every articles(upto 2 decimals)
            (a, an , the) 
            
            - create a function for the whole process
            
    
"""

def func(text):
    text = text.strip()
    
    dict_words = {}
    word = ""
    
    
    for c in text + " ":
        c = c.strip()
        if c.isalpha():
            word = word + c.lower()
        else:
            if word.isspace():
                word = ""
            else:
                dict_words[word] = 1 + dict_words.get(word,0)
                word = ""
                
    del dict_words[""]
    
    
    by_keys = sorted(dict_words.items())
    by_keys = dict(by_keys)
    by_values = sorted(dict_words.items(), key = lambda x:x[1] )
    by_values = dict(by_values)
    
    for c in by_keys:  #counting the no of occurences of the a, an, the....
        if c == "a":
            a_count = by_keys[c]
        if c == "an":
            an_count = by_keys[c]
        if c == "the":
            the_count = by_keys[c]
    
    words_count = by_keys.values() #creating the list of values of dictionary
    total = 0
    for i in words_count:  #counting total words...
        total = total + i
    
    a_percentage = a_count / total * 100 
    an_percentage = an_count / total * 100
    the_percentage = the_count / total * 100
    
    return by_keys,by_values,["{:.2f}".format(a_percentage),"{:.2f}".format(an_percentage),"{:.2f}".format(the_percentage)]
    
    
        
        
    
    
    
    
    
    
    

    
text = """Plato is widely considered a pivotal figure in the history of Ancient Greek and Western philosophy,
along with his teacher, Socrates, and his most famous student, Aristotle. He has often been cited as one of the
founders of Western religion and spirituality. The so-called neoplatonism of philosophers, such as Plotinus 
and Porphyry, greatly influenced Christianity through Church Fathers such as Augustine. Alfred North Whitehead 
once noted: "the safest general characterization of the European philosophical tradition is that it consists of
a series of footnotes to Plato."

Plato was an innovator of the written 
dialogue and dialectic forms in philosophy.
Plato is also considered the founder of Western
political philosophy. His most famous contribution
is the theory of Forms known by pure reason, 
in which Plato presents a solution to the problem of 
universals known as Platonism (also ambiguously called
either Platonic realism or Platonic idealism). 
He is also the namesake of Platonic love and the Platonic solids.

His own most decisive philosophical 
influences are usually thought to have been,
along with Socrates, the pre-Socratics Pythagoras,
Heraclitus and Parmenides, although few of his predecessors' 
works remain extant and much of what we know about these figures 
today derives from Plato himself. Unlike the work of nearly all 
of his contemporaries, Plato's entire body of work is believed to have 
survived intact for over 2,400 years. Although their popularity has 
fluctuated, Plato's works have consistently been read and studied.
"""
#cite:https://en.wikipedia.org/wiki/Plato

a,b,c = func(text)
print(a,"-> sorted by keys")
print()
print(b,"-> sorted by values")
print()
print(c,"-> percentage of the articles")
```
